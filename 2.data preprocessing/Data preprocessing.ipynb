```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset=pd.read_csv("loan_approval_dataset.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((dataset < lower_bound) | (dataset > upper_bound)).any(axis=1)
outliers = dataset[outliers]
```

```python
from scipy import stats
# Calculate Z-scores
z_scores = np.abs(stats.zscore(dataset.select_dtypes(include=[np.number])))

# Identify outliers (Z-score > 3 or < -3)
outliers = (z_scores > 3)

# Replace outliers with mean
df_clean = dataset.copy()
for column in dataset.select_dtypes(include=[np.number]).columns:
    mean_value = dataset[column].mean()
    df_clean.loc[outliers[column], column] = mean_value
    df_clean
```

```python
 df_clean
```

```python
print(f"Data after replacing outliers with mean using Z-score method:\n{df_clean}")
```
